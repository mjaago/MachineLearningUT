```python
import requests
import json
```

```python
to_cancel = []

with open('BIsToRefund.txt') as billing_infos:
    to_cancel = [line.strip() for line in billing_infos.readlines()]
```

```python
to_cancel
```

```python
url = 'http://epic.epinternal.net/billing/internal/payment/reservation/cancel'
myobj = '213709848'

for bi_id in to_cancel:
    try:
        x = requests.post(url, json = bi_id)
        status = json.loads(x.text)['billingInfoStatus']
        print(status)
    except:
        print('failed for ', bi_id)
```
